```python
import json
import pytz
import os
import asyncio
import pandas as pd
import numpy as np
import nest_asyncio

from telethon import TelegramClient
from datetime import datetime

pd.set_option('display.max_columns', None)


from telethon.sessions import MemorySession

client = TelegramClient(MemorySession(), api_id, api_hash)
```

```python
# Создаем папку для сохранения изображений
save_dir = '/Applications/Diploma/rian_images'
os.makedirs(save_dir, exist_ok=True)
```

```python
# Настройки путей
save_dir = '/Applications/Study/Diploma/images'
os.makedirs(save_dir, exist_ok=True)

json_path = '/Applications/Study/Diploma/news_data.json'


channels = [
    ["@rian_ru"],
    ["@sportrian", "@sportsru", "@UFCRussia", "@football_nik", "@hockey_vbros"],
    ["@rianovostiAmerica", "@politica_media", "@vv_volodin", "@margaritasimonyan"],
    ["@World_Sanctions"],
    ["@igmtv", "@inQsitor"],
    ["@movie7channel", "@marvel4"],
    ["@rasofficial"]
]

categories = ["General news", "Sport", "Politic", "Economic", "Games", "Films and serials", "Science"]
limit = 400

async def fetch_messages(client, ch_name, category):
    channel_entity = await client.get_entity(ch_name)
    messages = await client.get_messages(channel_entity, limit=limit)

    data = {
        "category": [],
        "channel": [],
        "message_id": [],
        "time": [],
        "text": [],
        "image_path": []
    }

    for msg in messages:
        if msg.photo:
            image_path = os.path.join(save_dir, f'{ch_name[1:]}_{msg.id}.jpg')
            await msg.download_media(file=image_path)

            data["category"].append(category)
            data["channel"].append(ch_name)
            data["message_id"].append(msg.id)
            data["time"].append(msg.date.strftime("%d.%m.%Y"))
            data["text"].append(msg.message)
            data["image_path"].append(image_path)

            print(f"Saved: {image_path}")

        else:
            data["category"].append(category)
            data["channel"].append(ch_name)
            data["message_id"].append(msg.id)
            data["time"].append(msg.date.strftime("%d.%m.%Y"))
            data["text"].append(msg.message)
            data["image_path"].append("")

            print(f"Saved: {""}")

    return data

async def gather_data():
    client = TelegramClient('session_name', api_id, api_hash)
    await client.start()

    all_data = {
        "category": [],
        "channel": [],
        "message_id": [],
        "time": [],
        "text": [],
        "image_path": []
    }

    for cat, ch_list in zip(categories, channels):
        for ch_name in ch_list:
            channel_data = await fetch_messages(client, ch_name, cat)

            for key in all_data:
                all_data[key].extend(channel_data[key])

    with open(json_path, 'w', encoding='utf-8') as f:
        json.dump(all_data, f, ensure_ascii=False, indent=4)

    print(f"Data successfully saved to {json_path}")

    await client.disconnect()

if __name__ == '__main__':
    nest_asyncio.apply()
    asyncio.run(gather_data())
```
